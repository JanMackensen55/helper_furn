```python
pay_101010 = []
pay_001010 = []
pay_002000 = []
for i in range(len(difference)):
    p1 = 0
    p2 = 0
    p3 = 0
    di = difference[i]
    for j in range(6):
        if di[j]>=-10:
            p1 += 10
        if di[j]>=-5:
            #p1 += 10
            p2 += 10
            p3 += 20
        if di[j] == 0:
            p1 += 10
            p2 += 10
    pay_101010.append(p1)
    pay_001010.append(p2)
    pay_002000.append(p3)
pay_101010.sort()
pay_001010.sort()
pay_002000.sort()
```
