```python
import pandas as pd
import csv
```

```python
data = '/home/cath/Code/furniture_experiment/data/full_experiment/data_furnituremanufacturing_2021-08-11_09-15.csv'
questions = 'questions.csv'
new_file = 'answers_subjects.csv'

df = pd.read_csv(data, sep=';', encoding='UTF-16 LE')
df_questions = pd.read_csv(questions, sep=';')

variables = df_questions[:]['Code'].values.tolist()
labels = df_questions[:]['Frage'].values.tolist()
answers = df.loc[1:, variables]

# write a csv file with the label of every question as header, 
# the 'true question' as first line after that
# and the answers of every subject per line
with open(new_file ,'w', encoding='UTF8') as f:
    writer = csv.writer(f, delimiter =';')
    writer.writerow(( ['Fragencodes:'] + variables))
    writer.writerow(( ['Fragen:'] + labels))
    
    # get the row of answers for every subject, which is all her answers to every question
    # a row is a series of all given answers
    # an index is the index of the subject
    for index, row in answers.iterrows():
        # in data we store all the 'real' answers
        data = []
        # get the value for every answer and map it to its true value
        for question in range(0, row.size):
            possible_answers = df_questions['Antwortart'][question].split(',')
            answer = row[question]
            
            if pd.isna(answer):
                data.append('No answer')
                continue
                
            if len(possible_answers) == 1:
                data.append(answer)
            else:
                data.append(possible_answers[int(answer) -1])
                
        writer.writerow(( ['VPN ' + str(index)] + data))
```
